```python
from json.decoder import scanstring

import numpy as np
import matplotlib.pyplot as plt

x_train = np.array([1.0, 2.0])
y_train = np.array([300.0, 500.0])

b = 100
k = 100

def hanshu(x, k, b):
    m = x.shape[0]
    y = np.zeros(m)
    for i in range(m):
        y[i] = k * x[i] + b
    return y

tem_hanshu = hanshu(x_train, k, b)

plt.plot(x_train, tem_hanshu, c='b', label='Our Prediction')
plt.scatter(x_train, y_train, marker='x', c='r', label="Actual Data")

plt.xlabel('Days')
plt.ylabel('BMI')
plt.title("beauty")
plt.legend()
plt.show()
```
